# Exercício 2

|Base de cálculo mensal (R$) | alíquota |
|----------------------------|----------|
|Até 2.259,20 |isento|
|De 2.259,21 até 2.826,65| 7,5%|
|De 2.826,66 até 3.751,05 |15%|
|De 3.751,06 até 4.664,68 |22,5%|
|Acima de 4.664,68 |27,5%|

```python
# Para fins de melhor visualização, o calculo do imposto de renda será definido em uma função

def calcImpostoRenda(salario):
    baseCalculo = salario
    imposto = 0
    if baseCalculo > 4664.68:
        imposto = imposto + (baseCalculo - 4664.68) * 0.275
        baseCalculo = 4664.68
    if baseCalculo > 3751.05:
        imposto = imposto + (baseCalculo - 3751.06) * 0.225
        baseCalculo = 3751.06
    if baseCalculo > 2826.64:
        imposto = imposto + (baseCalculo - 2826.65) * 0.15
        baseCalculo = 2826.65
    if baseCalculo > 2259.20:
        imposto = imposto + (baseCalculo - 2259.20) * 0.075
    return imposto

print(f"Quem ganha R$ 10.450,00 pagará R$ {calcImpostoRenda(10450):6.2f}")
print(f"Quem ganha R$ 50.000,00 pagará R$ {calcImpostoRenda(50000):6.2f}")
print(f"Quem ganha R$ 100.000,00 pagará R$ {calcImpostoRenda(100000):6.2f}")
print(f"Quem ganha R$ 500.000,00 pagará R$ {calcImpostoRenda(500000):6.2f}")
print(f"Quem ganha R$ 1.000.000,00 pagará R$ {calcImpostoRenda(1000000):6.2f}")
print(f"Quem ganha R$ 100.000.000,00 pagará R$ {calcImpostoRenda(100000000):6.2f}")
```

```python
def calcNovoImpostoRenda(salario):
    baseCalculo = salario
    imposto = 0
    if baseCalculo > 4664.68:
        imposto = imposto + (baseCalculo - 4664.68) * 0.04
        baseCalculo = 4664.68
    if baseCalculo > 3751.05:
        imposto = imposto + (baseCalculo - 3751.06) * 0.03
        baseCalculo = 3751.06
    if baseCalculo > 2826.64:
        imposto = imposto + (baseCalculo - 2826.65) * 0.02
        baseCalculo = 2826.65
    if baseCalculo > 2259.20:
        imposto = imposto + (baseCalculo - 2259.20) * 0.01
    return imposto

salario = float(input("Insira o salário: "))
imposto = calcNovoImpostoRenda(salario)
if imposto==0.0:
    print(f'Com o salário de R${salario:6.2f} o trabalhador está isento de pagar o Imposto de Renda')
else:
    print(f'Com o salário de R${salario:6.2f} o trabalhador terá de pagar R${imposto:6.2f} de imposto')
```

# Exercício 6

Armazene 3 números nas variáveis A, B e C<br>
Crie outra variável temporária<br>
Se A for menor que B:<br>
&emsp;Armazene A em Temp<br>
&emsp;Passe o valor de B para A<br>
&emsp;Passe o valor de Temp para B<br>
Se A for menor que C:<br>
&emsp;Armazene A em Temp<br>
&emsp;Passe o valor de C para A<br>
&emsp;Passe o valor de Temp para C<br>
Se B for menor que C:<br>
&emsp;Armazene B em Temp<br>
&emsp;Passe o valor de C para B<br>
&emsp;Passe o valor de Temp para C<br>

```python
#Demonstracao em python
A = int(input("Digite o valor A: "))
B = int(input("Digite o valor B: "))
C = int(input("Digite o valor C: "))
temp = int()

if A<B:
    temp = A
    A = B
    B = temp
if A<C:
    temp = A
    A = C
    C = temp
if B<C:
    temp = B
    B = C
    C = temp
print("Os valores depois da ordenacao sao:")
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```

# Exercício 7

Armazene 3 números nas variáveis A, B e C<br>
Crie outra variável temporária chamada **Maior** e atribua o valor -Infinito a ela<br>
Se A for maior que **Maior** e o módulo de A por 2 for igual a zero:<br>
&emsp;Armazene A em **Maior**<br>
Se B for maior que **Maior** e o módulo de B por 2 for igual a zero:<br>
&emsp;Armazene B em **Maior**<br>
Se C for maior que **Maior** e o módulo de C por 2 for igual a zero:<br>
&emsp;Armazene C em **Maior**<br>
Se **Maior** for igual a -infinito:<br>
&emsp;**Maior** = 0 <br>
Senão:<br>
&emsp;Retorne o valor de **Maior**<br>

```python
#Demonstracao em python
A = int(input("Digite o valor A: "))
B = int(input("Digite o valor B: "))
C = int(input("Digite o valor C: "))
maior = 0  # Considere o menor número possível como 0

if A>maior and A%2!=0:
    maior = A
if B>maior and B%2!=0:
    maior = B
if C>maior and C%2!=0:
    maior = C
print(f'O maior valor impar é: {maior}' if maior!=0 else 'Nao ha nenhum numero impar maior')
```

```python
num = int(input("Digite um número: "))
isInside = False
if num <= -2:
    print(f"o número {num} é menor ou igual a -2")
    isInside = True
if num >= -3 and num < -1:
    print(f"o número {num} é maior ou igual a -3 e menor ou igual a -1")
    isInside = True
if num > 0 and num < 2:
    print(f"o número {num} é maior que 0 e menor que 2")
    isInside = True
if num > 1 and num <= 3:
    print(f"o número {num} é maior que 1 e menor ou igual a 3")
    isInside = True
if num >= 4:
    print(f"o número {num} é maior ou igual a 4")
    isInside = True
if not(isInside):
    print(f"O número {num} não está em nenhum intervalo")
```
