# Exercício 8

```python
def isLeapYear(year):
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)

year = int(input("Digite o ano: "))
print(f'O ano {year} é bissexto' if isLeapYear(year) else f"O ano {year} não é bissexto")
```

# Exercício 9

```python
num = int(input("Digite um número: "))
isInside = False
if num <= -2:
    print(f"o número {num} é menor ou igual a -2")
    isInside = True
if num >= -3 and num < -1:
    print(f"o número {num} é maior ou igual a -3 e menor ou igual a -1")
    isInside = True
if num > 0 and num < 2:
    print(f"o número {num} é maior que 0 e menor que 2")
    isInside = True
if num > 1 and num <= 3:
    print(f"o número {num} é maior que 1 e menor ou igual a 3")
    isInside = True
if num >= 4:
    print(f"o número {num} é maior ou igual a 4")
    isInside = True
if not(isInside):
    print(f"O número {num} não está em nenhum intervalo")
```
