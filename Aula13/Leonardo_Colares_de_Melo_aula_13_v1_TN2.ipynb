## Exercício 01.3.1

```python
def exercicio_1_3_1(cores):
    lista_plana = []
    lista_de_listas = []
    
    for cor1 in cores:
        for cor2 in cores:
            if cor1 == cor2:
                lista_plana.append(cor1)
                lista_plana.append(cor2)
                lista_de_listas.append([cor1, cor2])
                
    return (lista_plana, lista_de_listas)
```

## Exercício 02.2.1

```python
def exercicio_2_2_1(lista_cores_objetos):
    dict_cor_chave = {}
    dict_objeto_chave = {}
    
    for cor, objeto in lista_cores_objetos:
        
        dict_cor_chave[cor] = objeto
        
        dict_objeto_chave[objeto] = cor
        
    return (dict_cor_chave, dict_objeto_chave)
```

## Exercício 02.2.2

```python
def exercicio_2_2_2(lista_cores_objetos):
    dict_cor_chave = {cor: obj for cor, obj in lista_cores_objetos}
    
    dict_objeto_chave = {obj: cor for cor, obj in lista_cores_objetos}
    
    return (dict_cor_chave, dict_objeto_chave)
```

## Exercício 03.3.1

```python
def exercicio_3_3_1(ini, fim):
    resultado = []
    limite_divisor = int(math.sqrt(fim))
    
    for n in range(ini, fim + 1):
        for d in range(2, limite_divisor + 1):
            if n % d == 0:
                resultado.append(n)
                break
                
    return resultado
```

## Exercício 03.3.2

```python
def exercicio_3_3_2(ini, fim):
    limite_divisor = int(math.sqrt(fim))
    
    return [n for n in range(ini, fim + 1) 
            if [d for d in range(2, limite_divisor + 1) if n % d == 0]]
```

## Exercício 03.4.1

```python
def exercicio_3_4_1(ini, fim):
    primos = []
    
    for n in range(ini, fim + 1):
        if n < 2: 
            continue
            
        eh_primo = True
        limite = int(math.sqrt(n))
        
        for d in range(2, limite + 1):
            if n % d == 0:
                eh_primo = False
                break
        
        if eh_primo:
            primos.append(n)
            
    return primos
```

## Exercício 03.4.2

```python
def exercicio_3_4_2(ini, fim):
    return [n for n in range(ini, fim + 1) 
            if n > 1 and not [d for d in range(2, int(math.sqrt(n)) + 1) if n % d == 0]]
```

## Exercício 03.5.1

```python
def exercicio_3_5_1(ini, fim):
    todos = set()
    nao_primos = set()
    
    for n in range(ini, fim + 1):
        if n < 2:
            continue
        
        todos.add(n)
        
        limite = int(math.sqrt(n))
        for d in range(2, limite + 1):
            if n % d == 0:
                nao_primos.add(n)
                break
    
    return todos - nao_primos
```

## Exercício 03.5.2

## Exercício 03.8.1

```python
def exercicio_3_8_1(texto: str) -> dict:
    tamanhos_palavras = {}
    
    palavras = texto.split()
    
    for palavra in palavras:
        tamanho = len(palavra)
        tamanhos_palavras[palavra] = tamanho
        
    return tamanhos_palavras
```

## Exercício 03.8.2

```python
def exercicio_3_8_2(texto: str) -> dict:
    tamanhos_palavras = {palavra: len(palavra) for palavra in texto.split()}
    
    return tamanhos_palavras
```

## Exercício 03.9.1

```python
def exercicio_3_9_1(ini: int, fim: int) -> list[int]:
    resultado = [
        next(d for d in range(9, 0, -1) if n % d == 0)
        for n in range(ini, fim + 1)
    ]
    
    return resultado
```

## Exercício 03.9.2

```python
def exercicio_3_9_2(ini: int, fim: int) -> dict[int, int]:
    resultado = {
        n: next(d for d in range(9, 0, -1) if n % d == 0)
        for n in range(ini, fim + 1)
    }
    
    return resultado
```
